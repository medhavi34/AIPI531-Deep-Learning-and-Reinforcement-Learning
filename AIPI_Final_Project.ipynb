```python
import pandas as pd
from collections import Counter

# Assuming the dataframes df_ratings and df_movies are already loaded as shown in the previous step
# Here's a simplified example of how to implement a most popular item recommender

# Function to calculate the most popular movies
def most_popular_movies(ratings_df, movies_df, top_n=10):
    """
    Function to find the most popular movies based on ratings count.

    Args:
    ratings_df (DataFrame): DataFrame containing ratings.
    movies_df (DataFrame): DataFrame containing movie details.
    top_n (int): Number of top movies to return.

    Returns:
    DataFrame: Top N most popular movies.
    """
    # Count the number of ratings for each movie
    movie_popularity = Counter(ratings_df['movie_id'])

    # Create a DataFrame from the counter
    df_movie_popularity = pd.DataFrame.from_dict(movie_popularity, orient='index', columns=['rating_count'])

    # Reset index to turn the indexes into a column
    df_movie_popularity.reset_index(inplace=True)

    # Rename the column to 'movie_id'
    df_movie_popularity.rename(columns={'index': 'movie_id'}, inplace=True)

    # Merge with the movies DataFrame to get movie titles
    df_popular_movies = pd.merge(df_movie_popularity, movies_df[['movie_id', 'title']], on='movie_id')

    # Sort by rating_count in descending order
    df_popular_movies.sort_values(by='rating_count', ascending=False, inplace=True)

    return df_popular_movies.head(top_n)

# Example usage
# top_10_popular_movies = most_popular_movies(df_ratings, df_movies, top_n=10)
# print(top_10_popular_movies)
```

```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can choose other models as well
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
def generate_movie_recommendation(user_input):
    # Tokenize the user input
    input_ids = tokenizer.encode(user_input, return_tensors="pt")

    # Generate movie recommendations
    with torch.no_grad():
        output = model.generate(input_ids, max_length=50, num_return_sequences=1, num_beams=5, no_repeat_ngram_size=2)

    # Decode the generated text
    recommendations = [tokenizer.decode(output[0], skip_special_tokens=True)]

    return recommendations
```

```python
while True:
    user_input = input("Enter your movie preferences (e.g., 'Recommend me a movie that is a comedy and has action scenes'): ")

    if user_input.lower() == "exit":
        break

    recommendations = generate_movie_recommendation(user_input)

    print("\nMovie Recommendations:")
    for recommendation in recommendations:
        print(recommendation)
```

```python
import pandas as pd
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can choose a different model if needed
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# URLs for the MovieLens 100K dataset
url_u_data = 'http://files.grouplens.org/datasets/movielens/ml-100k/u.data'
url_u_item = 'http://files.grouplens.org/datasets/movielens/ml-100k/u.item'

# Load user-item ratings (u.data)
columns_u_data = ['user_id', 'item_id', 'rating', 'timestamp']
u_data = pd.read_csv(url_u_data, sep='\t', names=columns_u_data)

# Load movie information (u.item)
columns_u_item = ['item_id', 'movie_title']
u_item = pd.read_csv(url_u_item, sep='|', encoding='latin-1', names=columns_u_item)

# Merge movie titles with user-item ratings
movie_data = pd.merge(u_data, u_item, on='item_id')

# Extract movie titles
movie_titles = movie_data["movie_title"].tolist()

# Combine movie titles to create prompts
movie_prompts = [f"Recommend me a movie titled {title}" for title in movie_titles]
```

```python
# Tokenize the movie prompts
tokenized_datasets = TextDataset(
    tokenizer=tokenizer,
    file_path=None,  # We're using in-memory data, so set this to None
    block_size=128,  # Adjust as needed
    overwrite_cache=True,
    examples=movie_prompts,
)

# Data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False,
)

# Training arguments
training_args = TrainingArguments(
    output_dir="./movie_recommendation_model",
    overwrite_output_dir=True,
    num_train_epochs=3,  # Adjust as needed
    per_device_train_batch_size=16,
    save_steps=10_000,
    save_total_limit=2,
)

# Initialize Trainer and start fine-tuning
trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=tokenized_datasets,
)

trainer.train()

# Save the fine-tuned model
trainer.save_model()
```

```python
import pandas as pd
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, DataCollatorForLanguageModeling, Trainer, TrainingArguments
from torch.utils.data import Dataset, DataLoader

# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Set the padding token to eos_token
tokenizer.pad_token = tokenizer.eos_token

# URLs for the MovieLens 100K dataset
url_u_data = 'http://files.grouplens.org/datasets/movielens/ml-100k/u.data'
url_u_item = 'http://files.grouplens.org/datasets/movielens/ml-100k/u.item'

# Load user-item ratings (u.data)
columns_u_data = ['user_id', 'item_id', 'rating', 'timestamp']
u_data = pd.read_csv(url_u_data, sep='\t', names=columns_u_data)

# Load movie information (u.item)
columns_u_item = ['item_id', 'movie_title']
u_item = pd.read_csv(url_u_item, sep='|', encoding='latin-1', names=columns_u_item)

# Merge movie titles with user-item ratings
movie_data = pd.merge(u_data, u_item, on='item_id')


movie_titles = movie_data["movie_title"].tolist()


user_preferences = ["I like action movies with strong characters", "I prefer romantic comedies", "I enjoy sci-fi with deep stories", "My favorite are horror films with psychological twists"]

# Create prompts that include user preferences
movie_prompts = []
for pref in user_preferences:
    for title in movie_data["movie_title"].tolist():
        prompt = f"Step 1: What features are most important to me when selecting movies (summarize my preferences briefly)? {pref} Step 2: Recommend me a movie titled {title}."
        movie_prompts.append(prompt)

tokenized_prompts = tokenizer(movie_prompts, return_tensors="pt", padding=True, truncation=True)

class MoviePromptsDataset(Dataset):
    def __init__(self, tokenized_prompts):
        self.tokenized_prompts = tokenized_prompts

    def __len__(self):
        return len(self.tokenized_prompts.input_ids)

    def __getitem__(self, idx):
        return {
            "input_ids": self.tokenized_prompts.input_ids[idx],
            "attention_mask": self.tokenized_prompts.attention_mask[idx],
        }

movie_prompts_dataset = MoviePromptsDataset(tokenized_prompts)
dataloader = DataLoader(movie_prompts_dataset, batch_size=16, shuffle=True)


data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=False,
)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

training_args = TrainingArguments(
    output_dir="./movie_recommendation_model",
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=16,
    save_steps=10_000,
    save_total_limit=2,
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=movie_prompts_dataset,
)

trainer.train()

# Save the fine-tuned model
trainer.save_model()
```
